### create dependent and independent variable

```python
# independent variables
x = df.drop(['CO2EMISSIONS'], axis= 1)

# dependent variable
y = df['CO2EMISSIONS']
```

### Create model

```python
from sklearn.linear_model import LinearRegression

# create empty model
model = LinearRegression()

# fit the model
model.fit(x,y)
```

### check model

```python
model.coef_
```

```python
model.intercept_
```

### save model to a file

```python
import pickle

# create empty binary file
model_file = open("co2_emission_prediction.pk", 'wb')

# dump model into empty file
pickle.dump(model, model_file)

# close file
model_file.close()
```
